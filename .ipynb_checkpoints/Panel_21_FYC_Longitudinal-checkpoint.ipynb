#### Why are there more observations for Panel 21 in the 2017 MEPS full-year consolidated file than in the MEPS longitudinal file for panel 21 (inscope during 2017)- 16,110 vs. 15,429?  See PROC FREQ output for details below.

##### The following is based 2017 MEPS Full-Year Consolidated File.

```sas
options nocenter;
title "MEPS 2017 Full-Year Consolidated File";
proc freq data=pufmeps.h201;
tables Panel;
run;
```

##### The following is based on the MEPS Longitudinal File (Panel 21, 2016-2017).

```sas
options nocenter nodate nonumber nosource;
ods html close;
proc format;
value yearind_fmt
1 = 'IN BOTH 2016 AND 2017'
2 = 'ONLY IN 2016'
3 = 'ONLY IN 2017';
 
value all5rds_fmt
  1 = 'Yes, Inscope/All 5 rounds'
  0 = 'No';
run;
```

```sas
title 'MEPS Longitudinal File (Panel 21 - 2016-2017)';
proc freq data=pufmeps.h202;
tables Panel*yearind*All5rds /list missing;
format yearind yearind_fmt.  all5rds  all5rds_fmt.;
run;
```
